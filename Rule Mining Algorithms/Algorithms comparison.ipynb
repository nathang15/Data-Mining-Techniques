```python
import numpy as np
import pandas as pd
import csv
from mlxtend.preprocessing import TransactionEncoder
import time
from pyECLAT import ECLAT
```

## APRIORI

```python
df = pd.read_csv(r"C:\\Users\\nhann\\Downloads\\basket_analysis.csv")
data = pd.DataFrame(df)
#small dataset
# data = df[:6]
# #large dataset
data = df
data
```

```python
# Mlxtend Apriori Function runtime
start_time = time.time()
from mlxtend.frequent_patterns import apriori
frequent_itemsets = apriori(data, min_support = 0.1, use_colnames=True)
# Sort frequent itemsets by support in descending order
frequent_itemsets = frequent_itemsets.sort_values(by='support', ascending=False)
print("--- Total runtime of Apriori implementation is: %s seconds ---" % (time.time() - start_time))
frequent_itemsets
```

## FP-GROWTH

```python
#Importing Libraries
from mlxtend.frequent_patterns import fpgrowth
#FP-growth algorithm runtime
start_time = time.time()
frequent_itemsets=fpgrowth(data,min_support=0.1, use_colnames=True)
# Sort frequent itemsets by support in descending order
frequent_itemsets = frequent_itemsets.sort_values(by='support', ascending=False)
print("--- Total runtime of Fp-Growth implementation is: %s seconds ---" % (time.time() - start_time))
frequent_itemsets
```

## ECLAT

```python
#process dataframe
df1 = pd.read_csv(r"C:\\Users\\nhann\\Downloads\\basket_analysis.csv")
df1_trans = df1.stack().reset_index()
df1_trans = df1_trans[df1_trans[0]] # filter out False values
df1_trans = df1_trans.drop(0, axis=1) # drop boolean column
df1_trans.columns = ['Transaction', 'Item'] # rename columns
df1_compressed = df1_trans.groupby('Transaction')['Item'].agg(lambda x: ','.join(sorted(set(x)))).reset_index()
df1_compressed.columns = ['Transaction', 'Items'] # rename columns
df1_compressed
data1 = pd.DataFrame(df1_compressed['Items'].str.split(',', expand=True))
data1
```

```python
from pyECLAT import ECLAT

start_time = time.time()
min_n_products = 1
min_support = 0.1
max_length = 13

# create an instance of eclat
my_eclat = ECLAT(data=data1, verbose=True)

# fit the algorithm
rule_indices, rule_supports = my_eclat.fit(min_support=min_support,
                                           min_combination=min_n_products,
                                           max_combination=max_length)
print("--- Total runtime of ECLAT is: %s seconds ---" % (time.time() - start_time))
```

```python
result = pd.DataFrame(rule_supports.items(),columns=['Item', 'Support'])
result
result.sort_values(by='Support', ascending=False)
```
